```python
#making the graphs dynamic with the dropdown for choosing model

@app.callback(Output('fig', 'figure'),[Input('mod', "value")])

def make_graph(modd=mod[0]):
            if modd=='Catboost':
                import os
                k = os.getcwd()
                k1 = k + '\\Catb.csv'
                ada = pd.read_csv(k1)
                true = ada['y_true'].values.tolist()
                pred = ada['y_pred'].values.tolist()    
                from sklearn.metrics import roc_curve 
                from sklearn.metrics import auc
                classes = list(set(true))
                n_classes = len(classes)
                tr = label_binarize(true, classes=classes)
                pr = label_binarize(pred,classes=classes)

                fpr,tpr,roc_auc = {},{},{}

                fig = make_subplots(rows=1, cols=2,
                                            specs=[[{'type':'scatter'}, {'type':'scatter'}]],
                                            subplot_titles=('ROC curve','Precision-recall curve'))


                for i in range(n_classes):
                                fpr[i], tpr[i], _ = roc_curve(tr[:, i], pr[:, i])
                for i in range(n_classes):
                                fig.add_trace(go.Scatter(x=fpr[i], y=tpr[i], mode='markers+lines',name='ROC curve for class %s'%(i)),row=1,col=1)
                t1 = [0.2,0.4,0.6,0.8,1]                           
                fig.add_trace(go.Scatter(x=t1, y=t1,name='Reference line'),row=1,col=1)

                #precision recall curve
                precision = dict()
                recall = dict()
                average_precision = dict()
                for i in range(n_classes):
                    precision[i], recall[i], _ = precision_recall_curve(tr[:, i],pr[:, i])
                    average_precision[i] = average_precision_score(tr[:, i], pr[:, i])
                for i in range(n_classes):
                        fig.add_trace(go.Scatter(x=precision[i], y=recall[i], mode='markers+lines',name='Precision-recall curve for class %s with area %s'%(i,average_precision[i])),row=1,col=2)

                # Compute micro-average ROC curve and ROC area
                precision["micro"], recall["micro"], _ = precision_recall_curve(tr.ravel(),
                    pr.ravel())
                average_precision["micro"] = average_precision_score(tr, pr,
                                                                     average="micro")
                fig.add_trace(go.Scatter(x=precision['micro'], y=recall['micro'], mode='markers+lines',name='Micro-avg precision-recall curve  with area %s'%(average_precision['micro'])),row=1,col=2)
                # Update xaxis properties
                fig.update_xaxes(title_text="FPR", row=1, col=1,title_font_size=20)
                fig.update_xaxes(title_text="Recall", title_font_size=20,row=1, col=2)
               

                # Update yaxis properties
                fig.update_yaxes(title_text="TPR",title_font_size=20, row=1, col=1)
                fig.update_yaxes(title_text="Precision", title_font_size=20,row=1, col=2)
                


                fig.update_layout(width=2000,height=500, 
                     paper_bgcolor= 'rgb(255,255,255)', xaxis=dict(showgrid=True,zeroline= True))
                return fig
            elif modd=='Random Forest':
                import os
                k = os.getcwd()
                k1 = k + '\\RF.csv'
                ada = pd.read_csv(k1)
                true = ada['y_true'].values.tolist()
                pred = ada['y_pred'].values.tolist()    
                               
                from sklearn.metrics import roc_curve 
                from sklearn.metrics import auc
                classes = list(set(true))
                n_classes = len(classes)
                tr = label_binarize(true, classes=classes)
                pr = label_binarize(pred,classes=classes)

                fpr,tpr,roc_auc = {},{},{}

                fig = make_subplots(rows=1, cols=2,
                                            specs=[[{'type':'scatter'}, {'type':'scatter'}]],
                                            subplot_titles=('ROC curve','Precision-recall curve'))


                for i in range(n_classes):
                                fpr[i], tpr[i], _ = roc_curve(tr[:, i], pr[:, i])
                for i in range(n_classes):
                                fig.add_trace(go.Scatter(x=fpr[i], y=tpr[i], mode='markers+lines',name='ROC curve for class %s'%(i)),row=1,col=1)
                t1 = [0.2,0.4,0.6,0.8,1]                           
                fig.add_trace(go.Scatter(x=t1, y=t1,name='Reference line'),row=1,col=1)

                #precision recall curve
                precision = dict()
                recall = dict()
                average_precision = dict()
                for i in range(n_classes):
                    precision[i], recall[i], _ = precision_recall_curve(tr[:, i],pr[:, i])
                    average_precision[i] = average_precision_score(tr[:, i], pr[:, i])
                for i in range(n_classes):
                        fig.add_trace(go.Scatter(x=precision[i], y=recall[i], mode='markers+lines',name='Precision-recall curve for class %s with area %s'%(i,average_precision[i])),row=1,col=2)

                # Compute micro-average ROC curve and ROC area
                precision["micro"], recall["micro"], _ = precision_recall_curve(tr.ravel(),
                    pr.ravel())
                average_precision["micro"] = average_precision_score(tr, pr,
                                                                     average="micro")
                fig.add_trace(go.Scatter(x=precision['micro'], y=recall['micro'], mode='markers+lines',name='Micro-avg precision-recall curve  with area %s'%(average_precision['micro'])),row=1,col=2)



                fig.update_layout(width=2000,height=500,
                     paper_bgcolor= 'rgb(255,255,255)', xaxis=dict(showgrid=True,zeroline= True))
                return fig
            
            elif modd=='Light GBM':
                import os
                k = os.getcwd()
                k1 = k + '\\LightGBM.csv'
                ada = pd.read_csv(k1)
                true = ada['y_true'].values.tolist()
                pred = ada['y_pred'].values.tolist()    
                
                from sklearn.metrics import roc_curve 
                from sklearn.metrics import auc
                classes = list(set(true))
                n_classes = len(classes)
                tr = label_binarize(true, classes=classes)
                pr = label_binarize(pred,classes=classes)

                fpr,tpr,roc_auc = {},{},{}

                fig = make_subplots(rows=1, cols=2,
                                            specs=[[{'type':'scatter'}, {'type':'scatter'}]],
                                            subplot_titles=('ROC curve','Precision-recall curve'))


                for i in range(n_classes):
                                fpr[i], tpr[i], _ = roc_curve(tr[:, i], pr[:, i])
                for i in range(n_classes):
                                fig.add_trace(go.Scatter(x=fpr[i], y=tpr[i], mode='markers+lines',name='ROC curve for class %s'%(i)),row=1,col=1)
                t1 = [0.2,0.4,0.6,0.8,1]                           
                fig.add_trace(go.Scatter(x=t1, y=t1,name='Reference line'),row=1,col=1)

                #precision recall curve
                precision = dict()
                recall = dict()
                average_precision = dict()
                for i in range(n_classes):
                    precision[i], recall[i], _ = precision_recall_curve(tr[:, i],pr[:, i])
                    average_precision[i] = average_precision_score(tr[:, i], pr[:, i])
                for i in range(n_classes):
                        fig.add_trace(go.Scatter(x=precision[i], y=recall[i], mode='markers+lines',name='Precision-recall curve for class %s with area %s'%(i,average_precision[i])),row=1,col=2)

                # Compute micro-average ROC curve and ROC area
                precision["micro"], recall["micro"], _ = precision_recall_curve(tr.ravel(),
                    pr.ravel())
                average_precision["micro"] = average_precision_score(tr, pr,
                                                                     average="micro")
                fig.add_trace(go.Scatter(x=precision['micro'], y=recall['micro'], mode='markers+lines',name='Micro-avg precision-recall curve  with area %s'%(average_precision['micro'])),row=1,col=2)



                fig.update_layout(width=2000,height=500,
                     paper_bgcolor= 'rgb(255,255,255)', xaxis=dict(showgrid=True,zeroline= True))
                return fig
            
            elif modd=='Logistic Regression':
                import os
                k = os.getcwd()
                k1 = k + '\\Log.csv'
                ada = pd.read_csv(k1)
                true = ada['y_true'].values.tolist()
                pred = ada['y_pred'].values.tolist()      

                from sklearn.metrics import roc_curve 
                from sklearn.metrics import auc
                classes = list(set(true))
                n_classes = len(classes)
                tr = label_binarize(true, classes=classes)
                pr = label_binarize(pred,classes=classes)

                fpr,tpr,roc_auc = {},{},{}

                fig = make_subplots(rows=1, cols=2,
                                            specs=[[{'type':'scatter'}, {'type':'scatter'}]],
                                            subplot_titles=('ROC curve','Precision-recall curve'))


                for i in range(n_classes):
                                fpr[i], tpr[i], _ = roc_curve(tr[:, i], pr[:, i])
                for i in range(n_classes):
                                fig.add_trace(go.Scatter(x=fpr[i], y=tpr[i], mode='markers+lines',name='ROC curve for class %s'%(i)),row=1,col=1)
                t1 = [0.2,0.4,0.6,0.8,1]                           
                fig.add_trace(go.Scatter(x=t1, y=t1,name='Reference line'),row=1,col=1)

                #precision recall curve
                precision = dict()
                recall = dict()
                average_precision = dict()
                for i in range(n_classes):
                    precision[i], recall[i], _ = precision_recall_curve(tr[:, i],pr[:, i])
                    average_precision[i] = average_precision_score(tr[:, i], pr[:, i])
                for i in range(n_classes):
                        fig.add_trace(go.Scatter(x=precision[i], y=recall[i], mode='markers+lines',name='Precision-recall curve for class %s with area %s'%(i,average_precision[i])),row=1,col=2)

                # Compute micro-average ROC curve and ROC area
                precision["micro"], recall["micro"], _ = precision_recall_curve(tr.ravel(),
                    pr.ravel())
                average_precision["micro"] = average_precision_score(tr, pr,
                                                                     average="micro")
                fig.add_trace(go.Scatter(x=precision['micro'], y=recall['micro'], mode='markers+lines',name='Micro-avg precision-recall curve  with area %s'%(average_precision['micro'])),row=1,col=2)



                fig.update_layout(width=2000,height=500,
                     paper_bgcolor= 'rgb(255,255,255)', xaxis=dict(showgrid=True,zeroline= True))
                return fig
            elif modd=='All':
                import os
                k = os.getcwd()
                k1 = k + '\\Ada.csv'
                ada = pd.read_csv(k1)
                true = ada['y_true'].values.tolist()
                pred = ada['y_pred'].values.tolist()    
                
                from sklearn.metrics import roc_curve 
                from sklearn.metrics import auc
                classes = list(set(true))
                n_classes = len(classes)
                tr = label_binarize(true, classes=classes)
                pr = label_binarize(pred,classes=classes)

                fpr,tpr,roc_auc = {},{},{}

                fig = make_subplots(rows=1, cols=2,
                                            specs=[[{'type':'scatter'}, {'type':'scatter'}]],
                                            subplot_titles=('ROC curve for AdaBoost','Precision-recall curve for AdaBoost'))


                for i in range(n_classes):
                                fpr[i], tpr[i], _ = roc_curve(tr[:, i], pr[:, i])
                for i in range(n_classes):
                                fig.add_trace(go.Scatter(x=fpr[i], y=tpr[i], mode='markers+lines',name='ROC curve for class %s'%(i)),row=1,col=1)
                t1 = [0.2,0.4,0.6,0.8,1]                           
                fig.add_trace(go.Scatter(x=t1, y=t1,name='Reference line'),row=1,col=1)

                #precision recall curve
                precision = dict()
                recall = dict()
                average_precision = dict()
                for i in range(n_classes):
                    precision[i], recall[i], _ = precision_recall_curve(tr[:, i],pr[:, i])
                    average_precision[i] = average_precision_score(tr[:, i], pr[:, i])
                for i in range(n_classes):
                        fig.add_trace(go.Scatter(x=precision[i], y=recall[i], mode='markers+lines',name='Precision-recall curve for class %s with area %s'%(i,average_precision[i])),row=1,col=2)

                # Compute micro-average ROC curve and ROC area
                precision["micro"], recall["micro"], _ = precision_recall_curve(tr.ravel(),
                    pr.ravel())
                average_precision["micro"] = average_precision_score(tr, pr,
                                                                     average="micro")
                fig.add_trace(go.Scatter(x=precision['micro'], y=recall['micro'], mode='markers+lines',name='Micro-avg precision-recall curve  with area %s'%(average_precision['micro'])),row=1,col=2)



                fig.update_layout(width=2000,height=500,
                     paper_bgcolor= 'rgb(255,255,255)', xaxis=dict(showgrid=True,zeroline= True))
                return fig
            elif modd=='AdaBoost':
                import os
                k = os.getcwd()
                k1 = k + '\\Ada.csv'
                ada = pd.read_csv(k1)
                true = ada['y_true'].values.tolist()
                pred = ada['y_pred'].values.tolist()      
    
                from sklearn.metrics import roc_curve 
                from sklearn.metrics import auc
                classes = list(set(true))
                n_classes = len(classes)
                tr = label_binarize(true, classes=classes)
                pr = label_binarize(pred,classes=classes)

                fpr,tpr,roc_auc = {},{},{}

                fig = make_subplots(rows=1, cols=2,
                                            specs=[[{'type':'scatter'}, {'type':'scatter'}]],
                                            subplot_titles=('ROC curve','Precision-recall curve'))


                for i in range(n_classes):
                                fpr[i], tpr[i], _ = roc_curve(tr[:, i], pr[:, i])
                for i in range(n_classes):
                                fig.add_trace(go.Scatter(x=fpr[i], y=tpr[i], mode='markers+lines',name='ROC curve for class %s'%(i)),row=1,col=1)
                t1 = [0.2,0.4,0.6,0.8,1]                           
                fig.add_trace(go.Scatter(x=t1, y=t1,name='Reference line'),row=1,col=1)

                #precision recall curve
                precision = dict()
                recall = dict()
                average_precision = dict()
                for i in range(n_classes):
                    precision[i], recall[i], _ = precision_recall_curve(tr[:, i],pr[:, i])
                    average_precision[i] = average_precision_score(tr[:, i], pr[:, i])
                for i in range(n_classes):
                        fig.add_trace(go.Scatter(x=precision[i], y=recall[i], mode='markers+lines',name='Precision-recall curve for class %s with area %s'%(i,average_precision[i])),row=1,col=2)

                # Compute micro-average ROC curve and ROC area
                precision["micro"], recall["micro"], _ = precision_recall_curve(tr.ravel(),
                    pr.ravel())
                average_precision["micro"] = average_precision_score(tr, pr,
                                                                     average="micro")
                fig.add_trace(go.Scatter(x=precision['micro'], y=recall['micro'], mode='markers+lines',name='Micro-avg precision-recall curve  with area %s'%(average_precision['micro'])),row=1,col=2)



                fig.update_layout(width=2000,height=500,
                     paper_bgcolor= 'rgb(255,255,255)', xaxis=dict(showgrid=True,zeroline= True))
                return fig
            

```
